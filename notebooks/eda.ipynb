# Exploratory data analysis

### Load Data

```python
# Set the project root
import sys
from pathlib import Path
from matplotlib import pyplot as plt

if '__file__' in locals():
    project_root = Path(__file__).resolve().parent.parent
else:
    project_root = Path.cwd().parent

sys.path.insert(0, str(project_root))
```

```python
import pandas

dataset = "data/EN_Card_Data.csv"

en_card_data_path = Path.cwd().parent / dataset
en_card_data = pandas.read_csv(en_card_data_path)
print(f"Loaded {dataset}. {len(en_card_data)} rows")
```

```python
# Show all headers
en_card_data.columns.tolist()
print(en_card_data.columns.tolist())
```

```python
# Show the distribution of HP values
plt.figure(figsize=(10, 6))
plt.hist(en_card_data['HP'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of HP Values')
plt.xlabel('HP')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
